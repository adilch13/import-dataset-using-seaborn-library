```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
flower=sns.load_dataset("iris")

flower

```

```python
sns.lineplot(x="sepal_length",y="sepal_width",data=flower)
plt.title("flowers plot")
plt.figure(figsize=(2,8))
plt.show()
```
